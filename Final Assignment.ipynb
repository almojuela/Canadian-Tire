# Analysis

```python
# Import Pandas & Numpy
import pandas as pd
import numpy as np

# This will create a DataFrame that is filled with all the info from the CSV file
tweets = pd.read_csv("data/canadian-tire_cct490.csv", index_col=False)

# Here, we're just looking at all the column names in the file itself
# This will help determine which ones we should keep, and which ones we can remove
tweets.columns
```

```python
tweets['created_at'].head()
# I'm just trying to see what kind of text is shown in the 'Created At' column
# Since we are focusing on the date, we're going to have to figure out
    # a way to separate the time and date from each other
```

```python
# We are taking the 'created at' column and splitting the date and time
tweets['created_at'].str.split(expand = True)
# This then creates a column called 'date'
tweets['date'] = tweets['created_at'].str.split(expand = True)[0]
# Then we'll create a column that just looks at the year and the day
tweets['month'] = tweets['date'].str.slice(0, 7)
```

```python
# Now, we will be creating Dataframes that just focus on specific months
    # These months will be May, June, and December, and we'll look at both years

# May 2013
may_2013 = tweets[tweets['month'] == '2013-05']
may2013 = may_2013[['month', 'text', 'positive', 'negative']]
    ## this is to create a separate Dataframe for May 2013 with speific columns
may13_positive = may2013[may2013['positive'] > 0] 
    # This looks at all the positive comments
may13_negative = may2013[may2013['negative'] > 0] 
    # This looks at all the negative comments
```
